# 데이터 불러오기

```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('/content/drive/MyDrive/전주/travel_advisor_preprocessing.csv')
df.head()
```

# Data preprocessing

```python
reviews = list(map(str,df.filtered_texts.tolist()))
```

```python
data_word = []

for i in tqdm(range(len(reviews))):
        tokens = reviews[i].split()
        choice = []
        for word in tokens:
          choice.append(word)
        data_word.append(choice)
data_word[:10]
```

# LDA

```python
from gensim.test.utils import common_corpus, common_dictionary
from gensim import corpora
from gensim.models.ldamodel import LdaModel
import matplotlib.pyplot as plt
```

```python
dictionary = corpora.Dictionary(data_word)
```

```python
len(dictionary)
```

```python
corpus = [dictionary.doc2bow(text) for text in data_word]
```

```python
num_topics = 10
chunksize = 1000
passes = 10
iterations = 400
eval_every = None

temp = dictionary[0]
id2word = dictionary.id2token
```

```python
perplexity_values = []

for i in range(2,15):
    ldamodel =  LdaModel(corpus=corpus,id2word=id2word,chunksize=chunksize,alpha='auto',eta='auto',iterations=iterations,
                  num_topics=i,passes=passes,eval_every=eval_every)
    perplexity_values.append(ldamodel.log_perplexity(corpus))
```

```python
x = range(2,15)
plt.plot(x,perplexity_values)
plt.xlabel('number of topics')
plt.ylabel('perplexity score')
plt.show()
```

```python
model = LdaModel(corpus=corpus,id2word=id2word,chunksize=chunksize,alpha='auto',eta='auto',iterations=iterations,
                  num_topics=10,passes=passes,eval_every=eval_every)
```

```python
top_topics = model.top_topics(corpus)
```

```python
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)
```
